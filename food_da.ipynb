```python
import pandas as pd
```

```python
df = pd.read_csv('archive/ifood_df.csv')
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
df = df[df['Income'] < df['Income'].quantile(0.99)]
df = df[df['Age'] < 100]
```

```python
education_dist = df[['education_2n Cycle', 'education_Basic', 'education_Graduation',
                     'education_Master', 'education_PhD']].sum()
print(education_dist)
```

```python
marital_dist = df[['marital_Divorced', 'marital_Married', 'marital_Single',
                     'marital_Together', 'marital_Widow']].sum()
print(marital_dist)
```

```python
df['Income'].fillna(df['Income'].median(), inplace=True)
df['Income_missing'] = df['Income'].isnull().astype(int)
```

```python
print(df.columns.tolist())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=range(20, 101, 5), kde=True, color='skyblue')
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.xticks(range(20, 101, 5))
plt.show()
```

```python
df['Income'].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define custom bins (5K increments)
bins = range(0, 100001, 5000)  # 0 to 100,000 in 5K steps

plt.figure(figsize=(10, 6))
sns.histplot(df['Income'], bins=bins, kde=True, color='steelblue')
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Number of Customers")
plt.xticks(bins, rotation=45)
plt.xlim(0, 95000)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#Income Segment
def income_segment(income):
    if income < 30000:
        return "Low Income"
    elif income <= 65000:
        return "Mid Income"
    else:
        return "High Income"
```

```python
df['IncomeSegment'] = df['Income'].apply(income_segment)
```

```python
# Education profile
education_columns = [col for col in df.columns if col.startswith("education_")]
df[education_columns].sum().plot(kind='bar', title="Education Level Counts")

plt.xticks(rotation=45)
plt.xlabel("Education Level")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.show()
```

```python
# Marital status profile
marital_columns = [col for col in df.columns if col.startswith("marital_")]
df[marital_columns].sum().plot(kind='bar', title="Marital Status Counts")

plt.xticks(rotation=45)
plt.xlabel("Education Level")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.show()
```

```python
sns.histplot(df['MntTotal'], bins=30, kde=True)
plt.title("Total Customer Spend Distribution")
```

```python
df.groupby('IncomeSegment')['MntTotal'].mean().plot(kind='bar')
plt.title('Average Spend by Income Segment')
plt.ylabel('Total Spend')

plt.xticks(rotation=45)
plt.xlabel("Education Level")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.show()
```

```python
product_cols = [
    'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds'
]

df[product_cols].mean().sort_values(ascending=False)
```

```python
df.groupby('IncomeSegment')[product_cols].mean().T.sort_values(
    by=['Low Income', 'Mid Income', 'High Income'], ascending=False
)
```

```python
df['FavoriteProduct'] = df[product_cols].idxmax(axis=1)
ct = pd.crosstab(df['IncomeSegment'], df['FavoriteProduct'])
```

```python
ct.plot(kind='bar', figsize=(10, 6))
plt.title("Favorite Product by Income Segment")
plt.xticks(rotation=45)
plt.xlabel("Income Segment")
plt.ylabel("Number of Customers")
plt.legend(title="Favorite Product", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Kidhome', y='Income', data=df, estimator='mean', ci=None)
plt.title("Average Income by Number of Kids at Home")
plt.xlabel("Number of Kids at Home")
plt.ylabel("Average Income")
plt.grid(axis='y')
plt.show()
```

```python
df['TotalChildren'] = df['Kidhome'] + df['Teenhome']
print(df['Income'].corr(df['TotalChildren']))
```

```python
plt.figure(figsize=(8, 5))
sns.regplot(x='TotalChildren', y='Income', data=df, scatter_kws={'alpha':0.6})
plt.title("Income vs. Number of Kids at Home")
plt.xlabel("Number of Kids at Home")
plt.ylabel("Income")
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Income', y='NumWebVisitsMonth', data=df,
            scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
plt.title("Income Distribution by Number of Web Visits per Month")
plt.ylabel("Web Visits Per Month")
plt.xlabel("Income")
# plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.show()
```

```python
correlation = df['Income'].corr(df['NumWebVisitsMonth'])
print(f"Correlation between Income and NumWebVisitsMonth: {correlation:.3f}")
```
